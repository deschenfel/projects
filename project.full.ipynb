```python
#H₀ (nula): As médias das avaliações dos usuários para Xbox One e PC são iguais
#H₁ (alternativa): As médias das avaliações dos usuários para Xbox One e PC são diferentes
from scipy import stats

# Filtrar os dados do período
df_periodo = df[(df['year_of_release'] >= 2014) & (df['year_of_release'] <= 2016)]

# Filtrar plataformas Xbox One e PC dentro do período, removendo user_scores ausentes
xone_scores = df_periodo[(df_periodo['platform'] == 'XOne') & (df_periodo['user_score'].notna())]['user_score']
pc_scores = df_periodo[(df_periodo['platform'] == 'PC') & (df_periodo['user_score'].notna())]['user_score']

# Aplicar o teste t (Welch)
t_stat, p_value = stats.ttest_ind(xone_scores, pc_scores, equal_var=False)

# Mostrar os resultados
print(f'Estatística t: {t_stat:.2f}')
print(f'Valor-p: {p_value:.4f}')
```

Como o valor-p (0.1160) é maior que 0.05, não temos evidências suficientes para rejeitar a hipótese nula. Aceitamos H₀ → As médias das avaliações dos usuários para Xbox One e PC podem ser consideradas estatisticamente iguais nesse período.

```python
# α = 0,05 (5%)
from scipy import stats

# Filtrar os dados do período
df_periodo = df[(df['year_of_release'] >= 2014) & (df['year_of_release'] <= 2016)]

# Filtrar dados dos gêneros Action e Sports
action_scores = df_periodo[(df_periodo['genre'] == 'Action') & (df_periodo['user_score'].notna())]['user_score']
sports_scores = df_periodo[(df_periodo['genre'] == 'Sports') & (df_periodo['user_score'].notna())]['user_score']

# Teste t de Welch (sem assumir variâncias iguais)
t_stat, p_value = stats.ttest_ind(action_scores, sports_scores, equal_var=False)

print(f'Estatística t: {t_stat:.2f}')
print(f'Valor-p: {p_value:.4f}')
```

Como p < 0.05, rejeitamos a hipótese nula. Há diferença estatisticamente significativa entre as médias de avaliação dos usuários para os gêneros Action e Sports nesse período. A estatística t está 8.88, muito distante do zero, ou seja, maior a diferen

Teste bicaudal que aprendemos; Padrão para teste t de comparação de médias entre dois grupos.
5% de chance de rejeitar a hipótese nula quando ela é verdadeira. Portanto, 95% de confiança nos resultados.
