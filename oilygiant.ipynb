```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
def load_data(file_path):
    """Carrega os dados de um arquivo CSV."""
    return pd.read_csv(file_path)

def train_and_evaluate_model(data):
    """Treina e avalia o modelo de regressão linear."""
    # Dividir os dados em características e alvo
    X = data[['f0', 'f1', 'f2']]
    y = data['product']
    
    # Dividir os dados em conjunto de treinamento e validação (75:25)
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=0)
    
    # Treinar o modelo
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Fazer predições
    predictions = model.predict(X_val)
    
    # Calcular o volume médio previsto de reservas
    average_predicted = np.mean(predictions)
    
    # Calcular o REQM (Root Mean Squared Error)
    reqm = np.sqrt(mean_squared_error(y_val, predictions))
    
    # Salvar predições e respostas corretas em um DataFrame
    results = pd.DataFrame({'Actual': y_val, 'Predicted': predictions})
    
    return average_predicted, reqm, results

def analyze_results(results):
    """Analisa os resultados e imprime as informações relevantes."""
    print("Resultados da Análise:")
    print(results.head())
    print("\nVolume médio previsto de reservas:", results['Predicted'].mean())
    print("REQM do modelo:", np.sqrt(mean_squared_error(results['Actual'], results['Predicted'])))
```

```python
# Lista de arquivos
files = ['geo_data_0.csv', 'geo_data_1.csv', 'geo_data_2.csv']

for file in files:
    print(f"\nAnalisando {file}:")
    data = load_data(file)
    average_predicted, reqm, results = train_and_evaluate_model(data)
    analyze_results(results)
```

```python
def calculate_profit(data, price_per_barrel=4.5, num_wells=200):
    """
    Calcula o lucro potencial dos poços de petróleo com base nas predições do modelo.

    :param data: DataFrame contendo as predições e os volumes de reservas reais.
    :param price_per_barrel: Preço por barril de petróleo.
    :param num_wells: Número de poços a serem considerados para o cálculo.
    :return: lucro potencial e as predições dos poços selecionados.
    """
    # Selecionar os 200 poços com os volumes previstos mais altos
    top_wells = data.nlargest(num_wells, 'Predicted')
    
    # Calcular o lucro potencial
    total_volume = top_wells['Predicted'].sum() * 1000  # Convertendo para unidades (barris)
    total_revenue = total_volume * price_per_barrel
    total_cost = 100_000_000  # Custo total do investimento
    profit = total_revenue - total_cost
    
    return profit, top_wells
```

```python
# Resultado para cada região
results = {}

# Lista de arquivos e médias de volume
files = ['geo_data_0.csv', 'geo_data_1.csv', 'geo_data_2.csv']
average_volumes = {
    'geo_data_0': 92.27,
    'geo_data_1': 69.15,
    'geo_data_2': 94.71
}

# Analisando cada região
for file in files:
    # Carregar os dados
    data = load_data(file)
    
    # Treinar o modelo e obter predições
    _, _, results_df = train_and_evaluate_model(data)
    
    # Calcular lucro potencial
    profit, top_wells = calculate_profit(results_df)
    
    # Armazenar resultados
    results[file] = {
        'profit': profit,
        'top_wells': top_wells
    }
```

```python

# Analisando os resultados
for file, result in results.items():
    print(f"\nResultados para {file}:")
    print(f"Lucro potencial: ${result['profit']:.2f}")
    print("Top 200 poços selecionados:")
    print(result['top_wells'][['Actual', 'Predicted']].head())

# Conclusão e recomendação
best_region = max(results.items(), key=lambda x: x[1]['profit'])
print(f"\nRecomendação: Desenvolver poços na região {best_region[0]} com lucro potencial de ${best_region[1]['profit']:.2f}.")
```

```python

import numpy as np

def bootstrap_profit_distribution(top_wells, price_per_barrel=4.5, num_samples=1000):
    """
    Realiza bootstrapping para calcular a distribuição de lucros.

    :param top_wells: DataFrame com os 200 poços selecionados.
    :param price_per_barrel: Preço por barril de petróleo.
    :param num_samples: Número de amostras para o bootstrapping.
    :return: lucro médio, intervalo de confiança de 95% e probabilidade de prejuízo.
    """
    profits = []
    
    # Cálculo do lucro real para os 200 poços
    total_volume = top_wells['Predicted'].sum() * 1000  # Convertendo para unidades (barris)
    total_revenue = total_volume * price_per_barrel
    total_cost = 100_000_000  # Custo total do investimento
    initial_profit = total_revenue - total_cost
    
    # Realizando bootstrapping
    for _ in range(num_samples):
        sample = np.random.choice(top_wells['Predicted'], size=len(top_wells), replace=True)
        total_volume_sample = sample.sum() * 1000
        total_revenue_sample = total_volume_sample * price_per_barrel
        profit_sample = total_revenue_sample - total_cost
        profits.append(profit_sample)
    
    # Cálculo do lucro médio e intervalo de confiança
    mean_profit = np.mean(profits)
    confidence_interval = np.percentile(profits, [2.5, 97.5])
    
    # Cálculo da probabilidade de prejuízo
    risk_of_loss = np.mean(np.array(profits) < 0) * 100  # Expressando como porcentagem
    
    return mean_profit, confidence_interval, risk_of_loss
```

```python
# Resultado para cada região com bootstrapping
bootstrap_results = {}

for file, result in results.items():
    mean_profit, confidence_interval, risk_of_loss = bootstrap_profit_distribution(result['top_wells'])
    bootstrap_results[file] = {
        'mean_profit': mean_profit,
        'confidence_interval': confidence_interval,
        'risk_of_loss': risk_of_loss
    }
```

```python

# Analisando os resultados de bootstrapping
for file, stats in bootstrap_results.items():
    print(f"\nResultados de Bootstrapping para {file}:")
    print(f"Lucro médio: ${stats['mean_profit']:.2f}")
    print(f"Intervalo de Confiança de 95%: (${stats['confidence_interval'][0]:.2f}, ${stats['confidence_interval'][1]:.2f})")
    print(f"Risco de Prejuízo: {stats['risk_of_loss']:.2f}%")

# Conclusão e recomendação
best_region_bootstrap = max(bootstrap_results.items(), key=lambda x: x[1]['mean_profit'])
print(f"\nRecomendação: Desenvolver poços na região {best_region_bootstrap[0]} com lucro médio de ${best_region_bootstrap[1]['mean_profit']:.2f}.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

PRICE_PER_BARREL = 4.5
BUDGET = 100_000_000
N_STUDY = 500      # pontos estudados por rodada
N_SELECT = 200     # poços desenvolvidos
N_BOOT = 1000
RANDOM_STATE = 0   # para reprodutibilidade

def train_and_get_validation_frame(df, random_state=RANDOM_STATE):
    """Treina regressão linear e retorna um DataFrame de validação com product (real) e pred (previsto)."""
    X = df[['f0','f1','f2']]
    y = df['product']
    X_tr, X_val, y_tr, y_val = train_test_split(
        X, y, test_size=0.25, random_state=random_state
    )
    model = LinearRegression().fit(X_tr, y_tr)
    preds = model.predict(X_val)

    val_df = X_val.copy()
    val_df['product'] = y_val.values       # alvo real
    val_df['pred'] = preds                 # previsão do modelo
    return val_df

def profit_from_selected(selected_df, price_per_barrel=PRICE_PER_BARREL, budget=BUDGET):
    """
    Calcula lucro usando o product REAL dos poços selecionados.
    product está em milhares de barris -> multiplicar por 1000 e pelo preço por barril.
    """
    revenue = selected_df['product'].sum() * 1000 * price_per_barrel
    return revenue - budget

def bootstrap_region(val_df, 
                     n_boot=N_BOOT, 
                     n_study=N_STUDY, 
                     n_select=N_SELECT, 
                     seed=RANDOM_STATE):
    """
    Em cada rodada:
      1) amostra 500 pontos (com reposição) do conjunto de validação;
      2) escolhe top-200 por 'pred';
      3) calcula lucro com 'product' REAL desses 200.
    Retorna lucro médio, IC 95% e risco de prejuízo (%).
    """
    rng = np.random.default_rng(seed)
    profits = np.empty(n_boot)

    for i in range(n_boot):
        # amostra 500 com reposição
        sample_idx = rng.choice(val_df.index, size=n_study, replace=True)
        sample = val_df.loc[sample_idx]

        # escolhe top-200 por previsão
        top = sample.nlargest(n_select, 'pred')

        # lucro com product real
        profits[i] = profit_from_selected(top)

    mean_profit = profits.mean()
    ci_low, ci_high = np.percentile(profits, [2.5, 97.5])
    risk_of_loss = (profits < 0).mean() * 100.0
    return mean_profit, (ci_low, ci_high), risk_of_loss, profits

# --- Carrega dados (ajuste caminhos se necessário) ---
geo0 = pd.read_csv('geo_data_0.csv')
geo1 = pd.read_csv('geo_data_1.csv')
geo2 = pd.read_csv('geo_data_2.csv')

# --- Treina e prepara frames de validação com real e previsto ---
val0 = train_and_get_validation_frame(geo0)
val1 = train_and_get_validation_frame(geo1)
val2 = train_and_get_validation_frame(geo2)

# --- Bootstrapping por região ---
stats = {}
for name, val in [('geo_data_0', val0), ('geo_data_1', val1), ('geo_data_2', val2)]:
    mean_p, (lo, hi), risk, profits = bootstrap_region(val)
    stats[name] = {
        'mean_profit': mean_p,
        'ci95': (lo, hi),
        'risk_loss_pct': risk
    }

# --- Impressão dos resultados ---
for name, s in stats.items():
    print(f"\n{name}:")
    print(f"  Lucro médio: ${s['mean_profit']:,.2f}")
    print(f"  IC 95%: (${s['ci95'][0]:,.2f}, ${s['ci95'][1]:,.2f})")
    print(f"  Risco de prejuízo: {s['risk_loss_pct']:.2f}%")

# --- Decisão final: filtra regiões com risco < 2.5% e escolhe a de maior lucro médio ---
candidatas = {k:v for k,v in stats.items() if v['risk_loss_pct'] < 2.5}
if candidatas:
    melhor = max(candidatas.items(), key=lambda kv: kv[1]['mean_profit'])
    regiao_escolhida, s = melhor
    print(f"\nRecomendação FINAL: desenvolver na região {regiao_escolhida} "
          f"(lucro médio ${s['mean_profit']:,.2f}, IC95% {s['ci95']}, "
          f"risco {s['risk_loss_pct']:.2f}%).")
else:
    # se nenhuma atende o critério de risco, reporta a melhor no trade-off
    melhor = max(stats.items(), key=lambda kv: kv[1]['mean_profit'])
    regiao_escolhida, s = melhor
    print("\nNenhuma região ficou abaixo de 2,5% de risco. "
          f"Ainda assim, a melhor relação retorno/risco é {regiao_escolhida} "
          f"(lucro médio ${s['mean_profit']:,.2f}, IC95% {s['ci95']}, "
          f"risco {s['risk_loss_pct']:.2f}%).")
```
